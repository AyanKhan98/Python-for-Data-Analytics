### How many UNIQUE country names are there ?

```python
#write your code
country=data['CountryName']
unique=[]
for i in country:
    if i not in unique:
        unique.append(i)
print("Total number of unique countries=",len(unique))
```

### Are there same number of country codes ?

```python
# How many unique country codes are there ? (should be the same #)
#write your code
code=data['CountryCode']
uniquecode=[]
for i in code:
    if i not in uniquecode:
        uniquecode.append(i)
print("Total number of unique country codes=",len(uniquecode))
```

### Are there many indicators or few ?

```python
# How many unique indicators are there ? (should be the same #)
#write your code
indicator=data['IndicatorCode']
uniqueind=[]
for i in indicator:
    if i not in uniqueind:
        uniqueind.append(i)
print("Total number of unique indicators=",len(uniqueind))
```

### Lets pick a country and an indicator to explore: CO2 Emissions per capita and the USA

```python
# select CO2 emissions for the United States
for i in range(len(data)+1):
    if 'United States' in data.iloc[i,0] and 'CO2 emissions' in data.iloc[i,2]:
        print(data.iloc[i,5])
```

```python
#see the dataset sliced for CountryCode = USA and IndicatorName = CO2 emmisions
for i in range(len(data)):
    if 'USA' in data.iloc[i,1] and 'CO2 emissions' in data.iloc[i,2]:
        print(data.iloc[[i]])
```

### Let's see how emissions have changed over time using MatplotLib

```python
# get the years
yr=[]
y={}
for i in year:
    if i not in yr:
        yr.append(int(i))
        y[int(i)]=0
# get the values for CO2 Emmisions 
for j in range(len(data)):
    if 'CO2 emissions' in data.iloc[j,2]:
        y[data.iloc[j,4]]+=int(data.iloc[j,5])
val=[]
for i in yr:
    val.append(y[i])
# create the bar plot using plt.bar
plt.bar(yr,val,align='center',alpha=1)
plt.show()
```

Turns out emissions per capita have dropped a bit over time, but let's make this graphic a bit more appealing before we continue to explore it.

```python
# switch to a line plot
plt.plot(yr,val)
# Label the axes
plt.xlabel('Year')
plt.ylabel('CO2 emissions')
#label the figure
plt.title('CO2 emissions per capita')
plt.xlim(1960,2015)
# to make more honest, start they y axis at 0


```
